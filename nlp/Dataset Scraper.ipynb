```python
# Imports
from selenium import webdriver
import matplotlib.pyplot as plt
from scipy.misc import imread

#########################
#  THE BIOARXIV SCRAPER #
#########################

# start Chrome
options = webdriver.ChromeOptions()
options.add_argument('headless')

client = webdriver.Chrome('/home/becks/code/test-mls/chromedriver', chrome_options=options)
client.get('https://www.biorxiv.org/')

# Only harvest from these relevant topics
interested_topics = [
    'Biochemistry', 
    'Bioengineering', 
    'Bioinformatics',
    'Biophysics',
    'Cancer Biology',
    'Cell Biology',
    'Clinical Trials',
    'Developmental Biology',
    'Epidemiology'
    'Evolutionary Biology'
    'Genetics'
    'Genomics'
    'Immunology'
    'Microbiology'
    'Molecular Biology'
    'Neuroscience'
    'Pathology'
    'Pharmacology and Toxicology'
    'Physiology'
    'Synthetic Biology'
    'Systems Biology'
]

topics = client.find_element_by_link_text('Bioinformatics')

topics.click() # Navigate to the Bioinformatics articles page
#client.get_screenshot_as_file('tmp.png')
#im = imread('tmp.png')
#plt.imshow(im)
#plt.show()

# find the page scroller at the bottom of the page and find out how many pages there are
pages = client.find_element_by_xpath("//ul[@class='pager pager-items']")
page_links = [pl.get_attribute('href') for pl in pages.find_elements_by_tag_name('a')]
first_page = page_links[0]
base_link, total_page_nums = page_links[-1].split('page=')
base_link += 'page='
all_pages = []
for i in range(int(total_page_nums)):
    # make a list of all the pages we'll have to navigate to
    all_pages.append('%s%d'%(base_link, i))

# start the scraping -> for each topic page, get the list of articles (title and url) 
# before moving to the next page
all_articles = []
for l in range(len(all_pages)):
    link = all_pages[l]
    client.get(link)
    article_list = client.find_element_by_xpath("//div[@class='highwire-list']//ul")
    items = article_list.find_elements_by_tag_name('li')

    for item in items:
        a = item.find_element_by_tag_name('a')
        url = a.get_attribute('href')
        title = a.find_element_by_tag_name('span').text
        all_articles.append({'title': title, 'url': url})
    
    print('Finished scraping page %d out of %s'%(l, total_page_nums))
```

```python
print(len(all_articles))
```

```python
options = webdriver.ChromeOptions()
options.add_argument('headless')

client = webdriver.Chrome('/home/becks/code/test-mls/chromedriver', chrome_options=options)

for x in range(len(all_articles)):
    article = all_articles[x]
    if article.get('abstract', 0) == 0:
        # in case the thing crashes, can start it up again at the last point
        client.get(article['url'])
        abstract = client.find_element_by_xpath("//div[@id='abstract-1']//p").text
        article['abstract'] = abstract
        print('Finished scraping %d of %d'%(x+1, len(all_articles)))
```

```python
captured = [a for a in all_articles if a.get('abstract', 0) != 0]
for c in all_articles:
    c['tag'] = 'Bioinformatics'
    
print(len(captured), captured[-1])
```

```python
import json

with open('biorxiv_scraped.json', 'w') as fp:
    print('dumping %d articles'%len(captured))
    json.dump(captured, fp)
```
